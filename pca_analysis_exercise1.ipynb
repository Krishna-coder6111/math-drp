Connecting libraries and drive

```python
import numpy as np
from google.colab import drive
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

drive.mount('/content/drive')
```

# First dataset

Adding the numpy file

```python
data = np.load('/content/drive/My Drive/colab/drp/dataset.npy')
data.shape

plt.figure(figsize=(8, 6))
plt.scatter(data[:, 0], data[:, 1], c='blue', label='Original Dataset')
plt.title('Original Dataset')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

## PCA on dataset

pcaing

```python
pca = PCA(n_components=2)
reduced_data_pca = pca.fit_transform(data)
explained_variance = pca.explained_variance_ratio_

plt.bar(range(len(explained_variance)), explained_variance, alpha=0.5, align='center', label='Explained variance ratio')
plt.ylabel('Explained Variance')
plt.xlabel('Principal Components')
plt.legend(loc='best')
plt.tight_layout()
plt.show()

explained_variance
```

## Interpretation

The high explained variance (96.2%) by the first principal component suggests that most of the information in the dataset can be captured by just that component. In simpler terms, the dataset is largely spread along one direction, and the second direction doesn't contribute much to the information in the dataset.

# Circle data

creating circle data

```python
num_samples = 200
theta = np.linspace(0, 2*np.pi, num_samples)

x = np.cos(theta)
y = np.sin(theta)

circle_data = np.column_stack((x, y))

plt.figure()
plt.scatter(circle_data[:, 0], circle_data[:, 1])
plt.title("2D Circle Points")
plt.xlabel("X-coordinate")
plt.ylabel("Y-coordinate")
plt.axis("equal")
plt.show()

circle_data.shape
```

## PCA on circle data

pcaing circle data

```python
pca_circle = PCA(n_components=2)

reduced_data_pca_circle = pca_circle.fit_transform(circle_data)

explained_variance_circle = pca_circle.explained_variance_ratio_

plt.bar(range(len(explained_variance_circle)), explained_variance_circle, alpha=0.5, align='center', label='Explained variance ratio')
plt.ylabel('Explained Variance')
plt.xlabel('Principal Components')
plt.legend(loc='best')
plt.tight_layout()
plt.show()

explained_variance_circle
```

## Interpretation

For the circle dataset, both principal components contribute almost equally to the variance (50.5% and 49.5%). This is quite different from the first dataset, where the first principal component dominated the explained variance. In the case of the circle, the data is spread out evenly along both dimensions, so both principal components are needed to capture the structure of the data.



# Comparison

1. First Dataset: One principal component could capture most
of the variance (96.2%), indicating that the data was primarily spread along one direction.

2. Circle Dataset: Both principal components are nearly equally important in explaining the variance, signifying a uniform spread in all directions.
